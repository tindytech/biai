This notebook provides a quick sanity check of your environment to make sure it is working as expected. Recall that if you are using a local machine setup, you will have to start the Python interpreter or place the commands in scripts and execute them from the command line.  If you are using Colab or Binder, you can just execute the cells below and view the output.

```python
# Import Libraries for Environment Check
import sys
import scipy
import numpy
import matplotlib
import pandas
import sklearn

# Print Versions of Python and Imported Libraries
print('Python: {}'.format(sys.version))
print('scipy: {}'.format(scipy.__version__))
print('numpy: {}'.format(numpy.__version__))
print('matplotlib: {}'.format(matplotlib.__version__))
print('pandas: {}'.format(pandas.__version__))
print('sklearn: {}'.format(sklearn.__version__))
```

If everything is installed correctly, you should see the version numbers for python and all of the imported libraries.

The imports used in the environment check are:

* `sys` provides access to some variables and functions associated with the Python interpreter, e.g. `.version`

* `scipy` contains modules for optimization, linear algebra, integration, interpolation, and other tasks common in science and engineering.

* `numpy` includes support for large, multi-dimensional arrays and matrices, along with high-level math functions to operate on these arrays.

* `matplotlib` is a plotting library that exposes an object-oriented API for embedding plots into applications.

* `pandas` enables data manipulation and analysis by offering data structures and operations for manipulating numerical tables and time series.

* `sklearn` features various classification, regression and clustering algorithms including support-vector machines, random forests, k-means and more.
